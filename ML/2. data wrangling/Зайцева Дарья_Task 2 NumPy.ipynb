#### 43. Дан массив (w,h,3) (картинка) dtype=ubyte, посчитать количество различных цветов

```python
w, h = 10, 10
img = np.random.randint(0, 10, (w, h, 3)).astype(np.ubyte)
col = img[..., 0] * 256 ** 2 + img[..., 1] * 256 + img[..., 2]
print(len(np.unique(col)))
```
